As in our previous notebooks we now seek a quantitative understanding of this charge/discharge phenomena.

To that end we draw a careful schematic, derive the Differential Equation for the output voltage, and solve it given our knowledge of the input voltage.


```python
# LR schematic
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    
    d += (V0 := elm.SourceV().label('$V_0$').dot())

    d += (L := elm.Inductor2().right().label('L').dot().label('$V_1$', 'right'))
    d += elm.CurrentLabel(top=False, ofst=.5).at(L).label('$I$')
    
    d += (R := elm.Resistor().down().label('R').dot())
    d += elm.CurrentLabel(top=True, ofst=.7).at(R).label('$I$').down()
    
    d += elm.Line().tox(V0.start).dot
    
    d += elm.Ground()
```

The Inductor Law states that
$$
  V_0-V_1 = LI'
$$
while Ohm's Law provides
$$
  V_1 = IR
$$
Substitution of the second into the first reveals
$$
  V_0 - IR = LI'.
$$
We rearrange to find

$$
  \boxed{\tau I' + I = V_0/R} \hskip 0.25in\hbox{where}\hskip 0.25in \boxed{\tau= L/R}  \tag{DE}
$$

is our time constant. Consistent with our tinkercad experiment above, we suppose that $V_0$ starts at 0 and then switches to say 5 (as we will soon be driving this with an Arduino). To that end, we suppose that our current, $I$, starts from 0, and proceed to solve
$$
  \tau I'(t) + I(t) = 5/R, \hskip 0.25in I(0)=0   \tag{DE5}
$$

We have been here before. We simplify this somewhat by writing $I$ as a constant part minus a dynamic part, namely

$$
   I(t) = 5/R - u(t)
$$

When we plug this guess into (DE5) we arrive at a DE for u,

$$
  \tau u'(t) + u(t) = 0, \hskip 0.25in u(0) = 5/R   \tag{DEu}
$$

Now this poses the question of what function looks like a multiple of its derivative. We now know that the exponential function fits the bill. In particular,

$$
  (\exp(t))' = \exp(t)
$$

and so, via the chain rule,

$$
  (\exp(-t/\tau))' = -(1/\tau)\exp(-t/\tau)
$$

hence, $u(t)=(5/R)\exp(-t/\tau)$, solves (DEu) and so


$$
  \boxed{I(t) = (5/R)(1-\exp(-t/\tau))}   \tag{I1}
$$

solves (DE5). As it is more convenient to measure voltage, we can turn this $I(t)$ into the output voltage via Ohm's Law

$$
  \boxed{V_1(t) = 5(1-\exp(-t/\tau))}   \tag{V1}
$$

We turn to python to graph this function, and compare it to the charging phase of our tinkercad work.

```python
# graph the exact step responce of the LR circuit

import matplotlib.pyplot as plt
import numpy as np

R = 10
L = 100e-3
tau = 1e3*L/R   # convert to milliseconds for easier viewing

V0 = 5
t = np.linspace(0,100,100)      # milliseconds
V1 = V0 * (1-np.exp(-t/tau))

plt.plot(t, V1, 'b')

plt.xlabel('t  (milliseconds)')
plt.ylabel('$V_1$  (volts)')
plt.grid('on')
plt.title('Inductor Charging with $\\tau$ = %d  (ms)' %(round(tau)));
```

That explicit formula, (V1), for $V_1$ gives us ready means to build an inductance meter. We note at time $t=\tau$ that

$$
  V_1(\tau) = 5(1-\exp(-1)) = 5\cdot 0.632
$$

and so, if we switch our pulse on, keep track of time and run time until $V_1=5\cdot 0.632$ then that time is then $\tau$. Knowing $\tau$ and $R$ we simply solve for

$$
  \boxed{L = R\tau}
$$

To build this device with an Arduino, we recall that 5 volts corresponds to a numerical value of 1023, and so we should switch our input from high to low and proceed until $V_1>672$ as $1023\cdot 0.632=672$. In practice, we observe that the
measured output voltage does not actually reach 1023. As such we first record A0max before scaling it by 0.632. In addition, to increase our accuracy we average our estimated inductance over 20 runs.

This example was using an inductance of 5000 mH. Of course we must explore its performance over a range of inductances.

Here is how our meter performs over a range of inductances.

```python
# report accuracy of our inductance meter
IndTrue = np.array([1000, 2000, 3000, 4000, 5000])    # mH
IndMeas = np.array([997, 1948, 2898, 3851, 4804])   # mH

RelErr = 100.0*np.abs(IndTrue - IndMeas)/IndTrue

plt.plot(IndTrue, RelErr, 'ko-')
plt.grid('on')
plt.xlabel('True Inductance, $mH$')
plt.ylabel('Percent Error')
plt.title('Inductance Meter: $R=1\ k\Omega$');
```

We see that our meter does poorly on inductances larger than 2000 mH.
